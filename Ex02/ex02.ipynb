```r
# import csv files library
library("readr")
library("ggplot2")
library("tidyr")
library("dplyr")
```

```r
# read csv files
df1 <- read_csv("TIME_USE_24092022.csv", show_col_types = FALSE)

# Eliminate every País = África do Sul
df1 <- df1[df1$País != "África do Sul",]

# Manipular o dataset para obter apenas os dados de Homens que trabalham ou não
df1 <- df1[df1$Ocupação == "Outros" | df1$Ocupação == "Trabalho não remunerado",]
df1 <- df1[df1$Sexo == "Homens",]

# dois diagramas de extremos e quantis
ggplot(df1, aes(x = Ocupação, y = Tempo, fill = Ocupação)) + geom_boxplot() + geom_jitter(width = 0.2, height = 0.2, alpha = 0.2) + labs(title = "Homens", x = "Ocupação", y = "Tempo médio (minutos)")


```
